```python
#Checking maximum length of values in a column
def max_value(column):
  max = 0
  for j in range(len(column)):
    len_del = len(column[j])
    if max < len_del:
      max =len_del
  return max
```

```python
max_val = max_value(df['RecentDelays'])
print(max_val)
```

```python
#Creating 3 columns and filling ti with null values
import numpy as np
for i in range(max_val):
  col_name = "delays_"+str(i)
  df[[col_name]] = np.nan
```

```python
#Filling the columns with appropriate values
for i in range(len(df['RecentDelays'])):
  for j in range(len(df['RecentDelays'][i])):
      col_name = "delays_"+str(j)
      df.loc[i,col_name] = df['RecentDelays'][i][j]
```
